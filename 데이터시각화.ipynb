## 막대그래프 시각화하기

```python
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df.shape
```

### 선실 등급(class)에 따른 생존률

```python
import matplotlib.pyplot as plt

sns.countplot(data=df, x="class", hue="survived")
plt.title("Survival Count by Class")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

### 성별에 따른 생존률

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="sex", hue="survived")
plt.title("Survival Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="class", y="survived", hue="sex", order=["First", "Second", "Third"])
plt.title("Survival Rate by Class and Sex")
plt.ylabel("Survival Rate")
plt.xlabel("Class")
plt.ylim(0, 1)
plt.show()
```

###  탑승 지역(embarked)에 따른 생존률

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x="embarked", y="survived", order=["C", "Q", "S"])
plt.title("Survival Rate by Embarkation Port")
plt.ylabel("Survival Rate")
plt.xlabel("Embarked")
plt.ylim(0, 1)
plt.show()
```

## 주식 가격 추이 시각화

```python
# 증권데이터 라이브러리
!pip install finance-datareader
```

```python
import FinanceDataReader as fdr
```

```python
# 삼성과 애플의 주가 데이터 수집
# 애플 첫 데이터가 1980-12-12, 삼성 첫 데이터가 1998-01-09이므로 시작일은 삼성 기준으로 맞춰줍니다.
samsung = fdr.DataReader("005930")
samsung_first_date = str(samsung.index[0])[:10]  # 삼성의 첫번째 데이터의 날짜, 추후 변동 가능
apple = fdr.DataReader("aapl", samsung_first_date)
```

```python
samsung
```

```python
apple
```

```python
# 결측치 확인
samsung.isnull().sum()
```

```python
apple.isnull().sum()
```

```python
#apple은 결측치가 없지만, samsung은 결측치가 있어서 0으로 바꿔준다
samsung = samsung.fillna(0)
samsung
```

```python
# 월별 그룹화를 위한 'Year-Month' 컬럼 생성
samsung['YearMonth'] = samsung.index.to_period('M')
apple['YearMonth'] = apple.index.to_period('M')

# 월별 평균 종가 계산
samsung_monthly = samsung.groupby('YearMonth')['Close'].mean()
apple_monthly = apple.groupby('YearMonth')['Close'].mean()
```

```python
# 삼성과 apple 주가 시작지점 맞추기
common_start = max(samsung_monthly.index[0], apple_monthly.index[0])
common_end = min(samsung_monthly.index[-1], apple_monthly.index[-1])

samsung_monthly = samsung_monthly.loc[common_start:common_end]
apple_monthly = apple_monthly.loc[common_start:common_end]
```

```python
import pandas as pd
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=samsung_monthly.index.to_timestamp(),
    y=samsung_monthly.values,
    mode='lines',
    name='Samsung (KRW)'
))

fig.add_trace(go.Scatter(
    x=apple_monthly.index.to_timestamp(),
    y=apple_monthly.values,
    mode='lines',
    name='Apple (USD)'
))

fig.update_layout(
    title='[정규화 없음] 월별 평균 종가 비교 (삼성 vs Apple)',
    xaxis_title='날짜',
    yaxis_title='평균 종가',
    template='plotly_white'
)

fig.show()

```

### 정규화한 그래프

```python
# 정규화: 첫 월의 값을 기준으로 나눔
samsung_norm = samsung_monthly / samsung_monthly.iloc[0]
apple_norm = apple_monthly / apple_monthly.iloc[0]

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=samsung_norm.index.to_timestamp(),
    y=samsung_norm.values,
    mode='lines',
    name='Samsung (정규화)'
))

fig.add_trace(go.Scatter(
    x=apple_norm.index.to_timestamp(),
    y=apple_norm.values,
    mode='lines',
    name='Apple (정규화)'
))

fig.update_layout(
    title='[정규화] 월별 평균 종가 추세 비교 (삼성 vs Apple)',
    xaxis_title='날짜',
    yaxis_title='정규화된 종가 (시작=1)',
    template='plotly_white'
)

fig.show()
```

### FinanceDataReader를 사용해 USD/KRW 환율 데이터를 받아옴

```python
import FinanceDataReader as fdr
import pandas as pd
import plotly.graph_objects as go

# 데이터 불러오기
samsung = fdr.DataReader('005930', '2010')
apple = fdr.DataReader('AAPL', '2010')
usdkrw = fdr.DataReader('USD/KRW', '2010')   # 환율 데이터

# 연/월 정보 추가
samsung['YearMonth'] = samsung.index.to_period('M')
apple['YearMonth'] = apple.index.to_period('M')
usdkrw['YearMonth'] = usdkrw.index.to_period('M')

# 월별 평균 종가 계산
samsung_monthly = samsung.groupby('YearMonth')['Close'].mean()
apple_monthly = apple.groupby('YearMonth')['Close'].mean()
usdkrw_monthly = usdkrw.groupby('YearMonth')['Close'].mean()

# 공통 구간 맞추기
common_index = samsung_monthly.index.intersection(apple_monthly.index).intersection(usdkrw_monthly.index)
samsung_monthly = samsung_monthly.loc[common_index]
apple_monthly = apple_monthly.loc[common_index]
usdkrw_monthly = usdkrw_monthly.loc[common_index]

# Apple 주가를 원화로 변환
apple_monthly_krw = apple_monthly * usdkrw_monthly
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=samsung_monthly.index.to_timestamp(),
    y=samsung_monthly.values,
    mode='lines',
    name='Samsung (KRW)'
))

fig.add_trace(go.Scatter(
    x=apple_monthly_krw.index.to_timestamp(),
    y=apple_monthly_krw.values,
    mode='lines',
    name='Apple (KRW 환산)'
))

fig.update_layout(
    title='[원화 기준] 월별 평균 종가 비교',
    xaxis_title='날짜',
    yaxis_title='평균 종가 (KRW)',
    template='plotly_white'
)

fig.show()
```
