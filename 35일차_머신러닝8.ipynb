```python
# 결측치 제거 ,필요없는 컬럼 제거

train= train.drop(['occyp_type'],axis = 1)
test= test.drop(['occyp_type'],axis = 1)
```

#Feature Engineering


## target 분리

```python
target = train.pop('credit')
```

```python
train.shape,test.shape
```

## 레이블 인코딩

```python
train
```

```python
## object형 컬럼은 'gender','car','reality'
cols= train.select_dtypes(include='object').columns # object형 컬럼 찾기
cols
```

```python
from sklearn.preprocessing import LabelEncoder

for col in cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])

train.head()
```

#모델링 및 하이퍼파라미터 튜닝


```python
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val = train_test_split(train,target,test_size=0.2,random_state=0)

X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

## 랜덤포레스트

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train,y_train)
pred = rf.predict(X_val) # 정확도, F1 score, 정밀도, 재현율
pred1 = rf.predict_proba(X_val) # roc_auc

print(rf.classes_)
pred[:10]
```

#평가 Metric 분석하기

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize

# 평가 지표
accuracy = accuracy_score(y_val, pred)
precision = precision_score(y_val, pred, average='weighted')
recall = recall_score(y_val, pred, average='weighted')
f1 = f1_score(y_val, pred, average='weighted')

print("정확도 (Accuracy):", accuracy)
print("정밀도 (Precision):", precision)
print("재현율 (Recall):", recall)
print("F1 점수 (F1 Score):", f1)

# ROC AUC
roc_auc = roc_auc_score(y_val, pred1, average='weighted', multi_class='ovr')
print("ROC AUC 점수:", roc_auc)
```
