````python
# 필요없는 컬럼 제거
train = train.drop(['transaction_id','jibun','apt','addr_kr','city','transaction_date'], axis=1)
test = test.drop(['transaction_id','jibun','apt','addr_kr','city','transaction_date'], axis=1)
````

LabelEncoder는 **학습한(label encoding된) 값 외의 새로운 값(test 데이터에만 있는 값)**을 만나면 에러를 발생시킨다


이상태로 레이블 인코딩을 할경우 train[col]에는 없던 '376-4' 같은 값이 test[col]에는 포함되어 있기 때문에 오류가 난다

````python
from sklearn.preprocessing import LabelEncoder
import pandas as pd


# train/test 병합을 위한 구분 컬럼 추가
train['is_train'] = 1
test['is_train'] = 0

# train/test 병합
full = pd.concat([train, test], ignore_index=True)

# object 타입 컬럼만 추출하여 Label Encoding 수행
cols = full.select_dtypes(include='object').columns

for col in cols:
    le = LabelEncoder()
    full[col] = le.fit_transform(full[col].astype(str))

# 다시 train/test 분리
train = full[full['is_train'] == 1].drop('is_train', axis=1).reset_index(drop=True)
test = full[full['is_train'] == 0].drop('is_train', axis=1).reset_index(drop=True)

````

#모델링 및 하이퍼파라미터 튜닝


````python
from sklearn.model_selection import train_test_split


# 데이터 분할
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

X_train.shape, X_val.shape, y_train.shape, y_val.shape
````

## 랜덤포레스트 회귀

````python
from sklearn.ensemble import RandomForestRegressor

# 모델 학습
rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, y_train)

# 예측 (예측 결과는 다시 지수 변환)
pred_log = rf.predict(X_val)
pred = np.expm1(pred_log)
y_val_true = np.expm1(y_val)
````

#평가 Metric 분석하기

````python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error


# 평가 지표 계산
mse = mean_squared_error(y_val_true, pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_val_true, pred)
r2 = r2_score(y_val_true, pred)

# 결과 출력
print(" 평가 지표")
print("MAE (Mean Absolute Error):", mae)
print("MSE (Mean Squared Error):", mse)
print("RMSE (Root Mean Squared Error):", rmse)
print("R² (R-squared):", r2)

# 예측값 일부 확인
print("\n 예측값 샘플:", pred[:10])
````
